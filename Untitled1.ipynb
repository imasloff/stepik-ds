```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

my_stat = pd.read_csv("https://stepik.org/media/attachments/course/4852/my_stat_1.csv")
```

```python
my_stat.session_value = my_stat.fillna({'session_value' : 0})
my_stat.loc[my_stat['n_users'] < 0, 'n_users'] = np.median(my_stat[my_stat['n_users'] >= 0].n_users)
```

```python
my_stat.groupby('group', as_index=False).agg({'session_value' : 'count'}).rename(columns={'session_value' : 'mean_session_value'})
```
